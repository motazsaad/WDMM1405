# Class Examples in Python 

## Person

```python
class Person:
    name = ''
    gender = ''
    age = 0 
    def __init__(self, n, g, a):
        self.name = n
        self.age = a
        self.gender = g
    def walk(self):
        print('I am walking')
    def eat(self):
        print('I am eating :)')
    def __str__(self):
        return self.name
```

```python
p1 = Person('Ahmed', 'm', 20)
p2 = Person('Sally', 'f', 18)
```

```python
p1.walk()
p2.eat()
```

```python
print(p1)
print(p2)
```

# Person Class (with None arguments in the constructor) 

```python
class Person:
    def __init__(self, name=None, age=0, gender=None):
        self.name = name 
        self.age = age
        self.gender = gender
    
    def walk(self):
        print('I am walking')
    
    def eat(self):
        print('I am eating :) ')
    def __str__(self): # string representation of the object 
        #return 'name: {}\tage: {}\tgender: {}'.format(self.name, self.age, self.gender)
        return '{}'.format(self.name)  
```

```python
p1 = Person()
p2 = Person(name='Ahmed')
p3 = Person(name='Hasan', age=20, gender='m')
```

```python
print(p1)
print(p2)
print(p3)
```

## Aria 



```python
class Circle:
    def __init__(self, c=(0,0), r=0):
        self.center = c 
        self.radius = r 
    def get_aria(self):
        return self.radius * self.radius * 3.14
    def get_perimeter(self):
        return 2 * self.radius * 3.14 
    def __str__(self):
        return 'center: {}\traduis:{}'.format(self.center, self.radius)
```

```python
c1 = Circle(c=(3,4), r=2)
c2 = Circle(c=(5,7), r=9)
```

```python
c1.get_aria()
```

```python
c2.get_perimeter()
```

```python
print(c1)
print(c2)
```
